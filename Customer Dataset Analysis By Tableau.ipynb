# Customer Dataset Analysis By Tableau

This is to show the data analysis work of Tableau.
<br>Explains will be followed by charts.

Data Source: https://www.kaggle.com/blastchar/telco-customer-churn 

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('/Users/samhou/Desktop/Python/TeleCustomerChurn/Basic View.jpg')
imgplot = plt.imshow(img)
fig = plt.gcf()
fig.set_size_inches(13,13)
plt
```

The chart shows the brief view of the dataset and the basic relationship from some metrics with churn. We get the number of whether people churn or not, the ratio of gender with churn, the ratio of senior citizen with churn and contract type with churn.
<br>For monthly contract, there is higher ratio of people who choose to churn.

```python
img = mpimg.imread('/Users/samhou/Desktop/Python/TeleCustomerChurn/Citizen + Churn VS Monthly Payment.jpg')
imgplot = plt.imshow(img)
fig = plt.gcf()
fig.set_size_inches(13,13)
plt
```

The chart shows the breakdown of senior citizen and churn with monthly charge in boxplot.
<br>People who are not senior citizens and no churn have wide range of monthly charges. And the oppisite situation, people who are senior citizens and churn have narrow range and higher monthly charges.

```python
img = mpimg.imread('/Users/samhou/Desktop/Python/TeleCustomerChurn/Payment Method VS Churn.jpg')
imgplot = plt.imshow(img)
fig = plt.gcf()
fig.set_size_inches(16,16)
plt
```

The chart shows the number of paperless billing and payment method with whether people churn or not. People who use electronic check and paperless billing have higher rate of churning. People who use mailed check tend not to churn, but their paperless billing using rate is lower than average.

```python
img = mpimg.imread('/Users/samhou/Desktop/Python/TeleCustomerChurn/Tech + Online Service VS Churn.jpg')
imgplot = plt.imshow(img)
fig = plt.gcf()
fig.set_size_inches(13,13)
plt
```

The chart shows the ralationship among tech support, online backup, online security and churn. For people who use all three service, they tend not to churn. For people who don't use any service, they tend to churn. And the group of people who may do online backup but no online security and tech support have higher ratio of churning.

```python
img = mpimg.imread('/Users/samhou/Desktop/Python/TeleCustomerChurn/Tenure VS Monthly Charges.jpg')
imgplot = plt.imshow(img)
fig = plt.gcf()
fig.set_size_inches(20,20)
plt
```

The chart shows the line chart of tenure VS average monthly charges. And the trend line is increasing. It means that people who have higher tenure may have higher monthly charges.
